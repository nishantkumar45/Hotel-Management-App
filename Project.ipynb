```python
# Global List/Dictionary Declaration 
Customers=[]
room_reserve=[]
room_details={}
total_rooms=[i for i in range(101,151)]
room_avaliable=total_rooms
```

```python
#class representing a Hotel customer
class Customer:
    #constuctor
    def __init__(self, name, PhoneNo, rooms_required, days,rooms_alloted):
        self.name = name
        self.PhoneNo = PhoneNo
        self.rooms_required = rooms_required
        self.days = days
        self.rooms_alloted= rooms_alloted

    def getDetails(self):
        return "%s has booked %i rooms for %i days." % (self.name, self.rooms_required, self.days)
    
    def __str__(self):
        return f"{self.name} booked {self.rooms_required} rooms {self.rooms_alloted} for {self.days} days."
```

```python
#Enquiry function menu driven
def enquiry():
    print("-"*25,"ENQUIRY","-"*25)
    print("1.Search by name of customer")
    print("2.Search by room number")
    print("3.Checking for room availability")
    print("4.Exit")
    print("\n")
    choice3 = int(input("Enter your choice: "))
    if choice3==1:
        name= input("Enter name of the customer:")
        for c in Customers:
            if c.name == name:
                print(c)
    elif choice3==2:
        room=int(input("Enter the room number:"))
        for c in Customers:
            if room in c.rooms_alloted:
                print(c)
    elif choice3==3:
        print(f"{len(room_avaliable)} rooms avaliable")
    elif choice3==4:
        pass
    else:
        print("Wrong Input")
```

```python
def new_book():
    print("Enter the Details for booking:")
    n = str(input("Enter Name: "))
    p = str(input("Enter PhoneNo.: "))
    r = int(input("Enter No.of rooms required:"))
    d = int(input("No.of days: "))
              
    # checks if any field is not empty
    if n!= " " and p!= " " and r!= " " and d!= " ":
        if  r<= len(room_avaliable):
            rooms=[room_avaliable[i] for i in range(0,r)]
            Cust = Customer(n,p,r,d,rooms)
            Customers.append(Cust)
            for i in rooms:
                room_avaliable.remove(i)
            print("\t***ROOM BOOKED SUCCESSFULLY***\n")
            print(Cust.getDetails())
        else: 
            print(f"Sorry {len(room_avaliable)} rooms only avaliable")
    else:
        print("\tName, PhoneNo. , Rooms Required & Days cannot be empty..!!")
```

```python
def booking():
    print("-"*25,"BOOKING","-"*25)
    print("1.No of rooms available")
    print("2.No of rooms reserved")
    print("3.New Booking")
    print("4.Exit")
    print("\n")
    choice2 = int(input("Enter your choice: "))
    if choice2==1:
        print(f"{len(room_avaliable)} rooms avaliable")
    elif choice2==2:
        print(f"{50-len(room_avaliable)} rooms reserved")
    elif choice2==3:
        new_book()
    elif choice2==4:
        pass
    else:
        print("Wrong Input")
```
